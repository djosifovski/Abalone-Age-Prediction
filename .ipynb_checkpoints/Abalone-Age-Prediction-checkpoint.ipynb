## IV. Modeling<a class="anchor" id="chapter4"></a>

### IV.1 Model Training<a class="anchor" id="section4_1"></a>

```python
def model(model, X, y, shuffle_split, scoring='r2'):
    """
    Function to fit a regressor to the training data
    """
    print(f'Trainig of model: {model.__class__.__name__}')
    
    model.fit(X, y)
    
    scores = cross_val_score(model, X, y, cv=shuffle_split, scoring=scoring)
    
    score = np.round(scores.mean(), 4)
    
    print(f'Parameters: {model.get_params()}')
    print(f'Mean cross-validated score: {score}')
    
    return score
```

```python
def grid_search(model, X, y, param_grid, shuffle_split, scoring='r2'):
    '''
    Function to tune hyper-parameters of models with grid search
    '''
    
    grid_search = GridSearchCV(model,
                               param_grid,
                               scoring,
                               n_jobs=-1,
                               cv=shuffle_split
                              ).fit(X, y)

    best_score = np.round(grid_search.best_score_, 4)

    print(f'Best estimator: {grid_search.best_estimator_}')
    print(f'Best mean cross-validated score: {best_score}')

    return grid_search.best_estimator_, best_score
```

```python
shuffle_split = ShuffleSplit(test_size=0.15, n_splits=10)
```

```python
%%time

n_neighbors = [2, 3, 4, 5, 6]
weights = ['uniform', 'distance']
leaf_size = np.arange(10,41,2)
param_grid_KNeighborsRegressor = {'n_neighbors': n_neighbors,
                                  'weights': weights,
                                  'leaf_size': leaf_size
                                 }

KNeighborsRegressor1, KNeighborsRegressor_score1 = grid_search(
    model=KNeighborsRegressor,
    X=X_train,
    y=y_train,
    param_grid=param_grid_KNeighborsRegressor,
    shuffle_split=shuffle_split
)
```
