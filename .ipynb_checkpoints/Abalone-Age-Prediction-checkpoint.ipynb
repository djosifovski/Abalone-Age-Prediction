### Data Cleaning

```python
abalone = abalone[(abalone != 0).all(axis=1)]
```

```python
abalone.describe()
```

### Train/Test Split

```python
train, test = train_test_split(abalone, test_size=0.15, random_state=0)
```

```python
abaloneLen = len(abalone.index)
trainLen = len(train.index)
testLen = len(test.index)

trainPercent = np.round(trainLen/abaloneLen*100, 3)
print(trainPercent)
print(np.round(100 - trainPercent, 3))

print(abaloneLen, trainLen, testLen)
```

```python
X_train, y_train = train.drop('Rings', axis=1), train['Rings']
X_test, y_test = test.drop('Rings', axis=1), test['Rings']
```

## Preprocessing

```python
numerical_transformer = Pipeline([
    ('min-max', MinMaxScaler()),
    ('pca', PCA())
])

preprocessor = ColumnTransformer([
    ('1hot', OneHotEncoder(), make_column_selector(dtype_include='object')),
    ('num', numerical_transformer, make_column_selector(dtype_include='number'))
])
```

## Modeling

### Model Training

```python
models = [
    KNeighborsRegressor(n_jobs=-1),
    Ridge(),
    SVR(),
    DecisionTreeRegressor(random_state=0),
    RandomForestRegressor(n_jobs=-1, random_state=0),
    GradientBoostingRegressor(random_state=0),
    XGBRegressor()
]
```

```python
clfs = []

for model in models:
    clf = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', model)])
    clfs.append(clf)
```

```python
model_names = [
    '      k-Neighbors Regressor',
    '           Ridge Regression',
    '   Support Vector Regressor',
    '    Decision Tree Regressor',
    '    Random Forest Regressor',
    'Gradient Boosting Regressor',
    '                    XGBoost'
]

for name, clf in zip(model_names, clfs):
    clf.fit(X_train, y_train)
    print(f"{name}: {cross_val_score(estimator=clf, X=X_train, y=y_train, cv=10, scoring='r2').mean():.4f}")
```

### Hyperparameter Optimization

```python
clfs_opt = []
```

```python
%%time
# KNeighborsRegressor

n_neighbors = [2, 3, 4, 5, 6]
weights = ['uniform', 'distance']
leaf_size = np.arange(10,41,2)
param_grid_KNeighborsRegressor = {'regressor__n_neighbors': n_neighbors,
                                  'regressor__weights': weights,
                                  'regressor__leaf_size': leaf_size
                                 }

clfs_opt.append(GridSearchCV(
    clfs[0],
    param_grid=param_grid_KNeighborsRegressor,
    scoring='r2',
    n_jobs=-1,
    cv=10
).fit(X_train, y_train))
```

```python
%%time
# Ridge

alpha = np.arange(0.00001, 1000, 5)
fit_intercept = [True, False]
solver = ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga', 'lbfgs']

param_grid_Ridge = {'regressor__alpha': alpha,
                    'regressor__fit_intercept': fit_intercept,
                    'regressor__solver': solver
                    }

clfs_opt.append(GridSearchCV(
    clfs[1],
    param_grid=param_grid_Ridge,
    scoring='r2',
    n_jobs=-1,
    cv=10
).fit(X_train, y_train))
```

```python
%%time

gamma = ['scale', 'auto']
C = [0.001, 0.01, 0.1, 1, 10, 100]

param_grid_SVR1 = {'regressor__kernel': ['linear'],
                   'regressor__C': C
                  }

# 'precomputed' is removed 
# reason: https://stackoverflow.com/questions/36306555/scikit-learn-grid-search-with-svm-regression

degree = [2, 3, 4]
param_grid_SVR2 = {'regressor__kernel': ['poly'],
                   'regressor__degree': degree,
                   'regressor__gamma': gamma,
                   'regressor__C': C
                  }


param_grid_SVR3 = {'regressor__kernel': ['rbf', 'sigmoid'],
                   'regressor__gamma': gamma,
                   'regressor__C': C
                  }

param_grid_SVR = [param_grid_SVR1, param_grid_SVR2, param_grid_SVR3]

clfs_opt.append(GridSearchCV(
    clfs[2],
    param_grid=param_grid_SVR,
    scoring='r2',
    n_jobs=-1,
    cv=10
).fit(X_train, y_train))
```

```python
%%time
# Decision Tree Regressor

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(1, 50, num = 11)]
max_depth.append(None)
# Number of features to consider at every split
max_features = ['auto', 'sqrt', 'log2']
max_features.append(None)
# Grow a tree with max_leaf_nodes in best-first fashion. Best nodes are defined as relative reduction in impurity.
max_leaf_nodes = np.arange(2, 30)
max_leaf_nodes = np.append(max_leaf_nodes, None)

param_grid_DecisionTreeRegressor = {'regressor__max_depth': max_depth,
                                    'regressor__max_features': max_features,
                                    'regressor__max_leaf_nodes': max_leaf_nodes
                                   }


clfs_opt.append(GridSearchCV(
    clfs[3],
    param_grid=param_grid_DecisionTreeRegressor,
    scoring='r2',
    n_jobs=-1,
    cv=10
).fit(X_train, y_train))
```

```python
%%time
# Random Forest Regressor

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 500, stop = 1000, num = 10)]
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(1, 100, num = 10)]
max_depth.append(None)
# Method of selecting samples for training each tree
bootstrap = [True, False]
    
param_grid_RandomForestRegressor = {'regressor__n_estimators': n_estimators,
                                    'regressor__max_depth': max_depth,
                                    'regressor__bootstrap': bootstrap
                                   }


clfs_opt.append(RandomizedSearchCV(
    clfs[4],
    param_distributions=param_grid_RandomForestRegressor,
    n_iter=20,
    scoring='r2',
    n_jobs=-1,
    cv=10,
    random_state=0
).fit(X_train, y_train))
```

```python
%%time
# Gradient Boosting Regressor

learning_rate = [0.001, 0.01, 0.1, 1, 10, 100]
max_depth = [int(x) for x in np.linspace(1, 10, num=5)]
min_samples_split = [int(x) for x in np.linspace(2, 10, num=5)]
n_estimators = [int(x) for x in np.linspace(start=100, stop=1000, num=5)]

param_grid_GradientBoostingRegressor = {'regressor__learning_rate': learning_rate,
                                        'regressor__max_depth': max_depth,
                                        'regressor__min_samples_split': min_samples_split,
                                        'regressor__n_estimators': n_estimators
                                       }

clfs_opt.append(RandomizedSearchCV(
    clfs[5],
    param_distributions=param_grid_GradientBoostingRegressor,
    n_iter=20,
    scoring='r2',
    n_jobs=-1,
    cv=10,
    random_state=0
).fit(X_train, y_train))
```

```python
%%time
# XGBoost

learning_rate = [0.01, 0.015, 0.025, 0.05, 0.1]
max_depth = [3, 5, 7, 9, 12, 15, 17, 25]
min_child_weigh = [1, 3, 5, 7, 9]

param_grid_XGBRegressor = {'regressor__learning_rate': learning_rate,
                           'regressor__max_depth': max_depth,
                           'regressor__min_child_weigh': min_child_weigh
                          }

clfs_opt.append(RandomizedSearchCV(
    clfs[6],
    param_distributions=param_grid_XGBRegressor,
    n_iter=20,
    scoring='r2',
    n_jobs=-1,
    cv=10,
    random_state=0
).fit(X_train, y_train))
```

### Model Evaluation

```python
MAEs = []
MSEs = []
RMSEs = [] # If True returns MSE value, if False returns RMSE value.
R2_scores = []
```

```python
for clf in clfs:
    MAEs.append(mean_absolute_error(y_test, clf.predict(X_test)))
    MSEs.append(mean_squared_error(y_test, clf.predict(X_test)))
    RMSEs.append(mean_squared_error(y_test, clf.predict(X_test), squared=False))
    R2_scores.append(r2_score(y_test, clf.predict(X_test)))
```

```python
MAEs_opt = []
MSEs_opt = []
RMSEs_opt = [] # If True returns MSE value, if False returns RMSE value.
R2_scores_opt = []
```

```python
for clf in clfs_opt:
    MAEs_opt.append(mean_absolute_error(y_test, clf.predict(X_test)))
    MSEs_opt.append(mean_squared_error(y_test, clf.predict(X_test)))
    RMSEs_opt.append(mean_squared_error(y_test, clf.predict(X_test), squared=False))
    R2_scores_opt.append(r2_score(y_test, clf.predict(X_test)))
```

### Summary of Scoring

```python
scoring_summary = pd.DataFrame({
    'Model': [name.strip() for name in model_names],
    'MAE score': MAEs,
    'MSE score': MSEs,
    'RMSE': RMSEs,
    'R2 Score': R2_scores
})
```

```python
scoring_summary.sort_values('R2 Score', ascending=False)
```

```python
scoring_summary_opt = pd.DataFrame({
    'Model': [name.strip() for name in model_names],
    'MAE score': MAEs_opt,
    'MSE': MSEs_opt,
    'RMSE': RMSEs_opt,
    'R2 Score': R2_scores_opt
})
```

```python
scoring_summary_opt.sort_values('R2 Score', ascending=False)
```
