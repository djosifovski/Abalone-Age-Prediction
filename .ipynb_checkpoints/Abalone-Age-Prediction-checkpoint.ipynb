##### 2. Standardization <a class="anchor" id="section_5_1_2"></a>

#### Notes:
- The values the featurea take are distributed close to a normal distribution, so it will probably be more beneficial to standardize the features (The Hundred-Page Machine Learning Book, Andriy Burkov, pg. 46).

```python
def standardization(df):
    """
    Function to standardize data
    """
    if df.min() < df.max(): # Skip constant value column
        df = (df - df.mean())/df.std()
    return df
```

```python
for col in X_train_num.columns:
    X_train_num.loc[:,col] = standardization(X_train_num.loc[:,col].copy()) #.copy() because of SettingWithCopyWarning
```

```python
for col in X_test_num.columns:
    X_test_num.loc[:,col] = standardization(X_test_num.loc[:,col].copy()) #.copy() because of SettingWithCopyWarning
```

```python
y_train = standardization(y_train.copy())
y_test = standardization(y_test.copy())
```

### VI. Model evaluation <a class="anchor" id="chapter6"></a>

#### Notes:
- To simplify this phase of the solution, three functions are implemented:
    - **train_regressor(model, X, y)**,
    - **cross_validation(model, X, y)**.

```python
def train_regressor(model, X, y):
    """
    Function to fit a regressor to the training data
    """
    print(f'Trainig of model: {model.__class__.__name__}')
    model.fit(X, y)
```

```python
def cross_validation(model, X, y):
    """
    Function to do ten-fold shuffle-splitcross-validation
    """
    shuffle_split = ShuffleSplit(test_size=0.5, train_size=0.5, n_splits=10)
    
    scores = cross_val_score(model, X, y, cv=shuffle_split, scoring='r2')
    
    scores = np.round(scores, 4)
    
    mean_scores = np.round(scores.mean(), 4)
    
    print(f'Cross-validation scores: {scores}')
    
    print(f'Mean cross-validated score: {mean_scores}')
    
    return mean_scores
```

```python
def model_evaluation(model, X, y):
    """
    Function to train and evaluate regressor
    """
    train_regressor(model, X, y)
    mean_score = cross_validation(model, X, y)
    
    return mean_score
```

### VII. Hyper-parameter tuning <a class="anchor" id="chapter7"></a>

#### Notes:
- Hyper-parameter tuning with Exhaustive Grid Search, and Ensemble learning algorithms.
- To simplify this phase of the solution, two functions:
    - **hyper_parameter_tuning(model, param_grid)**,
    - **evaluate_improvement(model, score_base, score_best)**.

```python
def hyper_parameter_tuning(model, param_grid, X, y):
    """
    Function to tune hyper-parameters with ten-fold cross-validation
    """
    shuffle_split = ShuffleSplit(test_size=0.5, train_size=0.5, n_splits=10)
    
    grid_search = GridSearchCV(model,
                               param_grid,
                               cv=shuffle_split,
                                ).fit(X, y)
    
    best_score = np.round(grid_search.best_score_, 4)
    
    print(f'Best estimator: {grid_search.best_estimator_}')
    print(f'Best mean cross-validated score: {best_score}')
    print(f'Best parameters: {grid_search.best_params_}')
    
    return grid_search.best_estimator_, best_score
```

```python
def evaluate_improvement(model, score_base, score_best):
    """
    Function to evaluate improvment of the model after hyper-parameter tuning
    """
    improvement = np.round((score_best - score_base)/score_base*100, 4)
    
    print(f'Base mean cross-validated score of model {model.__class__.__name__}: {score_base}')
    print(f'Improved mean cross-validated score of model {model.__class__.__name__}: {score_best}')
    print(f'Improvement of model {model.__class__.__name__}: {improvement}%')
    
    return improvement
```

```python
models.append(VotingRegressor.__class__.__name__)
models.append(KNeighborsRegressor.__class__.__name__ + ' - Baseline')
scores = [best_score_Ridge,
          best_score_SVR,
          best_score_SGDRegressor,
          best_score_GaussianProcessRegressor,
          best_score_DecisionTreeRegressor,
          best_score_RandomForestRegressor,
          best_score_GradientBoostingRegressor,
          score_VotingRegressor,
          score_KNeighborsRegressor]

model_scores_final = pd.DataFrame({'Model': models, 'Score': scores})
model_scores_final.sort_values('Score', ascending=False)
```

### VIII. Evaluation Metrics and Scoring <a class="anchor" id="chapter8"></a>

#### A. Predictions on validation dataset <a class="anchor" id="section_8_1"></a>

#### Notes:
- Regression metrics: $R^{2}$.

```python
def predict(model, X):
    """
    Function to predict target for samples in test set
    """
    print(f'Predicting target of model: {model.__class__.__name__}')
    y_pred = model.predict(X)
    return y_pred
```

```python
def metrics(model, X, y, y_pred):
    """
    Function to compute R2 regression score
    """
    R2_score = r2_score(y, y_pred)
    
    print(f'Evaluation of model: {model.__class__.__name__}')
    print(f'R2 score: {R2_score:.4f}')
    
    return R2_score
```
